Finding number of cat and dog images

```python
dog=0
cat=0

for img in files:
  name=img[0:3]
  if name=='dog':
    dog+=1
  else:
    cat+=1
print(f"No of Dog images {dog}")
print(f"No of Cat images {cat}")
```

Resizing all the images and put in a new directory

```python
os.mkdir('/content/image_resized')
```

```python
#Because we are using transfer learning we will take only first 2000 images and resize it and put in new folder
old_folder='/content/train/'
new_folder='/content/image_resized/'

for i in range(2000):
  img_name=files[i]
  img_path=old_folder+img_name

  img=Image.open(img_path)
  img=img.resize((224,224))
  img=img.convert('RGB')

  new_path=new_folder+img_name
  img.save(new_path)

```

```python
img = Image.open("/content/image_resized/dog.7088.jpg")
print(img.size)
```

Creating labels Dog=1, Cat=0

```python
labels=[]
for i in range(2000):
  img_name=files[i]
  animal=img_name[0:3]
  if animal=='dog':
    labels.append(1)
  else:
    labels.append(0)
```

```python
#Checking the labels for some initial images
for i in range(15):
  print(f'{files[i][:3]}=={labels[i]}')
print(len(labels))
```

```python
##Checking number of cats and dogs images in 2000
values, counts = np.unique(labels, return_counts=True)

print(values)
print(counts)
```

```python
import cv2
import glob
```

Converting the images in numpy array

```python
image_extension=['png','jpg']
#here all images are in .jpg format so it is not needed but in cases where both may be preent we want to try with both
file_list=[]

[file_list.extend(glob.glob(new_folder+'*.'+e)) for e in image_extension]

#cv2.imread return the particular image as numpy array. Now through np.asarray we are taking all those arry and putting in
# a single big array
dog_cat_images=np.asarray([cv2.imread(file) for file in file_list])
```

```python
print(dog_cat_images.shape)
print(type(dog_cat_images))
```

Training the pre-trained network

```python
X=dog_cat_images
y=np.asarray(labels)
#converting the labels list as array
```

Train test split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

**Scaling the data**

```python
X_train_scaled=X_train/255
X_test_scaled=X_test/255
```

```python
print(X_train_scaled[0])
#Doubt:-Why does it not return 3 channel 224x224 array on running above cmd
```

**Importing and training pre-trained model**

```python
!pip install tf-keras
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tf_keras as keras
```

```python
mobilenet_v2 ='https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'
# pretrained_model=hub.KerasLayer(mobilenet_v2,input_shape=(224,224,3),trainable=False)
```

```python
pretrained_model = hub.KerasLayer(
    mobilenet_v2,
    input_shape=(224,224,3),
    trainable=False
)
```

```python
no_of_classes=2
model=keras.Sequential([
    pretrained_model,
    keras.layers.Dense(no_of_classes)
])

# model = tf.keras.Model(inputs, outputs)
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['acc']
)
```

```python
history=model.fit(X_train_scaled,y_train,epochs=5)
```

```python
score,acc=model.evaluate(X_test_scaled,y_test)
print('Test Loss =', score)
print('Test Accuracy =', acc)
```

**Predictive system**

```python
img_input=input('Enter path of image')
img=cv2.imread(img_input)
cv2_imshow(img)
img_resize=cv2.resize(img,(224,224))
img_scaled=img_resize/255
img_reshape=np.reshape(img_scaled,[1,224,224,3])
prediction=model.predict(img_reshape)
label=np.argmax(prediction)

if(label==0):
  print("It is a Cat")
else:
  print("It is a Dog")
```
